# Normalized data

value = (value - min) / (max - min)

A total of 12 metrics, 10 metrics should be maximize, and 2 minimized

in order to normalize the values in the matrix that should be maximized : value = 1 - [(value - min) / (max - min)]

```python
# Load normalized_data sheet from data.xlsx file



normalized_df = pd.read_excel('data.xlsx', sheet_name='normalized_data')

normalized_df.head()
```

# Ordered Weighted Averages Operator

### the weights given by the decision maker : 

w = (0.05, 0.03, 0.05, 0.05, 0.009, 0.09, 0.1, 0.14, 0.03, 0.02, 0.04, 0.06, 0.091, 0.04, 0.02, 0.18)


```python
weights = np.array([0.05, 0.13, 0.08, 0.05, 0.09, 0.07, 0.04, 0.05, 0.06, 0.05, 0.08, 0.04, 0.05, 0.16])


print("The sum of all the weights : ", end='')
print(sum(weights))
```

### Calcul des scores

```python

data_set = normalized_df.to_numpy()

# array to store the calculated weighted values
averages = np.zeros(data_set.shape[0])

# calcul
for i in range(len(data_set)):
    res = 0
    temp = np.zeros(data_set.shape[1])
    for j in range(1, data_set.shape[1]):
        temp[j] = data_set[i][j]
    sorted_temp = np.sort(temp)[::-1] # sort the array
    res = np.dot(sorted_temp, weights)
    averages[i]= res

```

## Ranking of the countries

```python
ranking = list()

temp_averages = averages
max_index = np.argmax(temp_averages)
ranking.append(max_index)

for i in range(len(data_set)-1):
    temp_averages[max_index] = -9999
    max_index = np.argmax(temp_averages)
    ranking.append(max_index)

for i in range(len(ranking)):
    print(f'Choice number {i+1} : {data_set[ranking[i]][0].upper()}')

```
