# Ordered Weighted Averages Operator

### the weights given by the decision maker : 

w = (0.03, 0.03, 0.07, 0.05, 0.009, 0.09, 0.12, 0.12, 0.03, 0.02, 0.04, 0.06, 0.091, 0.04, 0.02, 0.18)


```python
weights = np.array([0.03, 0.03, 0.07, 0.05, 0.009, 0.09, 0.12, 0.12, 0.03, 0.02, 0.04, 0.06, 0.091, 0.04, 0.02, 0.18])


print("The sum of all the weights : ", end='')
print(sum(weights))
```

### Calcul des scores

```python

data_set = normalized_df.to_numpy()


# array to store the calculated weighted values
averages = np.zeros(data_set.shape[0])


# calcul
for i in range(1, len(data_set)):
    res = 0
    for j in range(1, data_set.shape[1]-1):
        res += data_set[i][j] * weights[j-1]
    averages[i-1]= res
```
